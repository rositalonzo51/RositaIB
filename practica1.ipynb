```python
with open ('libro1.txt') as f:
    libro1 = f.read()
```

```python
num_pal = len(libro1.split(' '))
```

```python
print( " El libro tiene " + str (num_pal) + " palabras") 
```

```python
num_pal_unic = len(set(libro1.split(' ')))
```

```python
print("El libro1  tiene " + str(num_pal_unic) + " palabras únicas")
```

```python
df_ili = pd.DataFrame({'palabras': libro1.split(' '), "ocurrencia":1})
```

```python
df_ili.head()
```

```python
print(", ".join(df_ili.groupby('palabras').count().sort_values(by='ocurrencia', ascending=False).head(20).index.tolist()))
```
